```python
import matplotlib.pyplot as plt

import numpy as np
import scipy.stats as stats
```

```python
def CLT():              
    colour = "#348ABD"
        
    nsamples = 1000
    samplesize = 30
    unif = stats.uniform        
        
    means = dict()
    for i in range(nsamples):
        rns = unif.rvs(size=samplesize)
        means[i] = np.mean(rns)   
    
    rns = unif.rvs(size=nsamples)
    f1 = plt.figure(1)
    f1.suptitle("Averages: $\mu = 0.5, \sigma^2 = %.5f$"%(1/12.0))      
    plt.hist(rns, 100, color=colour)
    f1.show()
    
    smean = np.mean(means.values())
    svar = stats.tvar(means.values())  
    f2 = plt.figure(2)
    f2.suptitle("Averages: $\mu = %.3f, \sigma^2 = %.5f$"%(smean, svar))      
    plt.hist(means.values(), 100, color=colour)
    f2.show()
    
    plt.show()
```

```python
CLT()
```
